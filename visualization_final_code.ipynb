```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#test data
data1 = np.random.rand(8, 8)
data2 = np.random.rand(8, 8)
#d = #the dictionary
#tricks = d['var1'] #replace with whatever the key ends up being
#cards = d['var2'] #replace with whatever the key ends up being
#n = d['decks'] #replace with whatever the key ends up being
#order needs to be confirmed with the data
x_axis_labels = ['BBB','BBR','BRB','BRR','RBB','RBR','RRB','RRR'] #player 1
y_axis_labels = ['BBB','BBR','BRB','BRR','RBB','RBR','RRB','RRR'] #player 2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

sns.heatmap(data1, annot= True, annot_kws={"size": 7},linewidths=.5, fmt=".2f", cmap='flare', 
            ax=ax1, cbar=False, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
ax1.set_title('Win by Tricks')
ax1.set_xlabel('Player 1', fontsize=8)
ax1.set_ylabel('Player 2', fontsize=8)

sns.heatmap(data2, annot= True, annot_kws={"size": 7},linewidths=.5, fmt=".2f", cmap='flare', 
            ax=ax2, cbar=False, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
ax2.set_title('Win by Cards')
ax2.set_xlabel('Player 1', fontsize=8)
ax2.set_ylabel('Player 2', fontsize=8)

cbar_ax = fig.add_axes([.95, 0.11, 0.02, .77])
cb = fig.colorbar(ax1.collections[0], cax=cbar_ax)
cb.outline.set_linewidth(.2)

plt.subplots_adjust(wspace=0.25)
plt.suptitle('Probability of Player 1 Winning Penney Game (n=*)', y=1.07, fontsize = 15) #replace(n=*)>...n =' +n+')'
plt.show()
```
