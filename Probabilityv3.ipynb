```python
##NOTE: THIS IS A WORK IN PROGRESS. PLEASE DON'T YELL AT ME IF IT LOOKS BAD. I'll GET SUPER SAD.


import sqlite3

# Establish SQLite in-memory connection
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create table for R and B
cursor.execute("CREATE TABLE cards (letter CHAR(1));")

# Insert R and B into the cards table
cursor.executemany("INSERT INTO cards (letter) VALUES (?);", [('R',), ('B',)])

# Query to generate all possible combinations for Player 1 and Player 2
query = """
SELECT A.letter || B.letter || C.letter AS player1_combination,
       D.letter || E.letter || F.letter AS player2_combination
FROM cards A
CROSS JOIN cards B
CROSS JOIN cards C
CROSS JOIN cards D
CROSS JOIN cards E
CROSS JOIN cards F;
"""

# Execute query and fetch results
cursor.execute(query)
results = cursor.fetchall()

# Close the connection
conn.close()
```

```python
#### VARATION 1 PROBABILITY ####

import random

def card_game_simulation(deck, p1_seq, p2_seq):

    #Automatic Tie if both sequences are the same
    if p1_seq == p2_seq:
        return 0
        
    pile = []
    p1_pile = []
    p2_pile = []
    
    for card in deck:
        pile.append(card)
        
        # Check if Player 1's sequence appears
        if ''.join(pile[-len(p1_seq):]) == p1_seq:
            p1_pile.extend(pile)
            pile.clear()
        
        # Check if Player 2's sequence appears
        elif ''.join(pile[-len(p2_seq):]) == p2_seq:
            p2_pile.extend(pile)
            pile.clear()
    
    # Determines the winner by card amount
    if len(p1_pile) > len(p2_pile):
        return 1  # P1 wins
    elif len(p2_pile) > len(p1_pile):
        return 2  # P2 wins
    else:
        return 0  # Tie

# Function to simulate multiple games for a given pair of player sequences
def sim_combos_var1(p1_seq, p2_seq, num_games):
    p1_wins = 0
    p2_wins = 0
    ties = 0

    for _ in range(num_games):
        # Even 26 for each color
        deck = ['R'] * 26 + ['B'] * 26
        random.shuffle(deck)
        result = card_game_simulation(deck, p1_seq, p2_seq)
        
        if result == 1:
            p1_wins = p1_wins + 1
        elif result == 2:
            p2_wins = p2_wins + 1
        else:
            ties = ties + 1
    
    # Calculate probabilities
    total_games = p1_wins + p2_wins + ties
    p1_prob = p1_wins / total_games
    p2_prob = p2_wins / total_games
    tie_prob = ties / total_games
    
    return p1_prob, p2_prob, tie_prob

# All combinations
def sim_com(num_games):
    new_results = []
    
    for p1_seq, p2_seq in results:
        p1_prob, p2_prob, tie_prob = sim_combos_var1(p1_seq, p2_seq, num_games)
        new_results.append({
            "P1 Sequence": p1_seq,
            "P2 Sequence": p2_seq,
            "P1 Win Probability": p1_prob,
            "P2 Win Probability": p2_prob,
            "Tie Probability": tie_prob
        })

    return new_results

# Simulate 10000 games for each combination
num_games = 10000
var1results = sim_com(num_games)
```

```python
### VARIATION 2 PROBABILITY ###

def sim_game(player1_combo, player2_combo, deck):
    random.shuffle(deck)  # Shuffle the deck

    # If both combos are the same, Player 1 should most likely just win
    if player1_combo == player2_combo:
        return 'Player 1'
    
    deck_string = ''.join(deck)
    
    # P1 and P2 combinations
    player1_index = deck_string.find(player1_combo)
    player2_index = deck_string.find(player2_combo)
    
    # Return the winner based on who appears first
    if player1_index == -1 and player2_index == -1:
        #If no combo
        return None
    elif player1_index != -1 and (player2_index == -1 or player1_index < player2_index):
        return 'Player 1'
    else:
        return 'Player 2'

# Run simulations for each pair of combinations
def run_sim(results, num_simulations=10000):
    new_results = []
    deck = ['R'] * 26 + ['B'] * 26  # Create the deck
    
    for Var2p1_guess, Var2p2_guess in results:
        player1_wins = 0
        player2_wins = 0
        
        for _ in range(num_simulations):
            winner = sim_game(Var2p1_guess, Var2p2_guess, deck)
            if winner == 'Player 1':
                player1_wins = player1_wins + 1
            elif winner == 'Player 2':
                player2_wins = player2_wins + 1
        
        # Calculate probabilities
        Var2P1ProbWin = player1_wins / num_simulations
        Var2P2ProbWin = player2_wins / num_simulations

        new_results.append({
            'P1 Sequence': Var2p1_guess,
            'P2 Sequence': Var2p2_guess,
            'P1 Win Probability': Var2P1ProbWin,
            'P2 Win Probability': Var2P2ProbWin
        })
    
    return new_results

Var2results = run_sim(results)
```
